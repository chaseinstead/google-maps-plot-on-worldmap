figuring out how to manually create color gradients in order to visualize altitude on the map

```python
from ast import literal_eval as make_tuple
import matplotlib.pyplot as plt


timestamps = []
number = 0
for i in range(250):
    timestamps.append(number)
    number += 1


altitudes = []
number = 0
for i in range(250):
    altitudes.append(number)
    number += 10
    

def create_test_colors(altitudes):
    '''returns list colorcodes (g (and b))'''
    colorcode_list = []
    color = 250
    for point in altitudes:    
        colorcode_list.append(color)
        color -= 1
    return colorcode_list


def make_rgb_code(colors):
    '''returns rgb codes from list of colorcodes'''
    final_colorcodes = []
    for i in colors:
        new = make_tuple("({}, {}, {})".format(1.0, i / 255.0, i / 255.0))
        final_colorcodes.append(new)
    return final_colorcodes


def find_point_color(altitude, change_color_after):
    '''returns the colorcode for each altitude. these can be inserdes in g and b'''
    if altitude == 0:
        altitude = 1
    times_11 = altitude / change_color_after
    colorcode = 255.0 - times_11
    return colorcode


#colors = create_test_colors(altitudes)

change_color_after = max(altitudes) / 255.0
colors = [find_point_color(i, change_color_after) for i in altitudes]

final_colors = make_rgb_code(colors)

fig, ax = plt.subplots()
fig.set_size_inches(22,3)
ax.set(xlabel='time', ylabel='altitude (meters)',
       title='visualizing altitutde test graph')
plt.scatter(timestamps, altitudes, c=final_colors)
```
